# Team Analysis

```python
# Import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Team data

```python
# Load team data
teams = pd.read_csv('data/relevant_data/MTeams.csv')
teams.head()
```

```python
# Load conference data
conferences = pd.read_csv('data/relevant_data/MTeamConferences.csv')
conferences.head()
```

```python
# Load coach data
coaches = pd.read_csv('data/relevant_data/MTeamCoaches.csv')
coaches.head()
```

```python
# Load city data
cities = pd.read_csv('data/relevant_data/Cities.csv')
cities.head()
```

```python
# Merge team, conference, coach, and city data
team_data = teams.merge(conferences, on='TeamID', how='left') \
    .merge(coaches, on=['TeamID', 'Season'], how='left')

# Merge cities only if CityID is available in team_data
if 'CityID' in team_data.columns:
    team_data = team_data.merge(cities, on='CityID', how='left')

team_data.head(25)
```

```python
# Save the merged data to a new CSV file
team_data.to_csv('data/processed/team_data.csv', index=False)
```
